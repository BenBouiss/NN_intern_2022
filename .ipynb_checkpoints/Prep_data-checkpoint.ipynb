```python
Horiz_resol = 2 #km/index
XstartYZ = int(500 / Horiz_resol)
YstartXZ = int(40 / Horiz_resol)
time = len(T)

for t in range(time):
    if (t+1)%int(time/10) == 0:
        print('Starting {} / {}'.format(t+1, time))
    df = pd.DataFrame()
    Melt = (Data['meltRate'][t].to_dataframe())
    T_F = Data['thermalDriving'][t].to_dataframe()
    S_F = Data['halineDriving'][t].to_dataframe()
    iceDraft = Data['iceDraft'].to_dataframe()
    df['meltRate'] = Melt
    df['thermalDriving'] = T_F
    df['halineDriving'] = S_F
    df['iceDraft'] = iceDraft
    df['date'] = t
    df.reset_index(inplace=True)
    
    if t == 0:
        dfT = df
    else:
        dfT = pd.concat([dfT, df], ignore_index = True)

dfT.rename(columns={'nx' :'x', 'ny': 'y'}, inplace=True)
dfT['x'] = dfT['x'] * Horiz_resol
dfT['y'] = dfT['y'] * Horiz_resol
```
